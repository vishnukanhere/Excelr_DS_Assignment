```python
UId_data = data.pivot_table(index='UId', columns='BT', values='BR', aggfunc='sum').reset_index(drop=True)
UId_data
```

```python
UId_data.fillna(0, inplace=True)
UId_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
UId_sim = 1 - pairwise_distances(UId_data.values, metric='cosine')
UId_sim
```

```python
### dircectly cosine similarity also get find

from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(UId_data.values)
```
